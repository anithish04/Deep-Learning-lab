# EX-1(Matrix Multiplication)

```python
import numpy as np
A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
C=np.dot(A,B)
print("matrix A:",A)
print("matrix B:",B)
print("Matrix C(A*B):",C)
```

# EX-2(Eigen Decomposition)

```python
import numpy as np
A=np.array([[4,2],[1,3]])
eigenvalue,eigenvectors=np.linalg.eig(A)
print("Matrix A:",A)
print("Eigen Values:",eigenvalue)
print("Eigen vectors:",eigenvectors)
```

# EX-3(SVD)

```python
import numpy as np
A=np.array([[3,2,2],[2,3,-2]])
U,S,V=np.linalg.svd(A)
print("matrix A:\n",A)
print("U matrix:\n",U)
print("Singular values:\n",S)
print("V matrix:\n",V)
```

# EX-4(Determinant calculation)

```python
import numpy as np
A=np.array([[6,1],[2,3]])
det=np.linalg.det(A)
print("matrix A:\n",A)
print("Determinant of A:\n",det)
```

# EX-5(Norm Calculation) 

```python
import numpy as np
V=np.array([3,4])
L1_norm=np.linalg.norm(V,ord=1)
L2_norm=np.linalg.norm(V)
print("vector V:\n",V)
print("L1 norm of V:\n",L1_norm)
print("L2 norm of V:\n",L2_norm)
```
